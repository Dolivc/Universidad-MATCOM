Carlos un apasionado de los cocteles cubanos a vivido toda su vida en La Habana. Ha tenido la suerte de probar una gran variedad de cocteles, pero siempre ha tenido preferencia por cuatro bebidas clasicas : la piña colada, el cuba libre, el mojito y el daiquiri. Inspirado por su amor a estas bebidas, Carlos se propone realizar un analisis en los distintos bares y restaurantes de La Habana. Su objetivo es descubrir si existen diferencias significativas en los precios de estos cocteles segun la ubicacion, el tipo de establecimiento y el tipo de coctel. Con su muestra de 231 restaurantes y bares, Carlos espera recopilar informacion valiosa y utilizar estos datos para responder sus preguntas y compartir sus hallazgos con otros amantes de los cocteles cubanos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from bibliotheque import data
```

```python
from bibliotheque import conteo_rest_gen
dfy = conteo_rest_gen(data)
fig = px.bar(dfy, x = "Municipio", y = "Contador" , color = "Propiedad", color_discrete_map={"Particular": "mediumvioletred", "Estatal": "blueviolet", "Mipyme": "blue", "Proyecto de desarrollo local": "forestgreen", "ONG":"cyan", "Cooperativa": "green"})
fig.update_traces(marker_line_color =["Black"], marker_line_width = 1)
fig.show()
```

Para profundizar en su analisis de los 231 restaurantes, Carlos decidio examinar los datos segun el tipo de propiedad y el municipio para comprender mejor las tendencias de los locales. Sus hallazgos revelaron que los restaurantes particulares son los que mas predominan en la ciudad, y que el municipio Playa tiene la mayor concentracion de este tipo de establecimientos. Esto se debe en parte a la popularidad creciente de las paladares en la gastronomia cubana. Ademas Carlos observo que el muinicipio de La Habana Vieja tiene una mayor proporcion de restaurantes estatales en comparacion con otros municipios. Pudo encontrar otros tipos de establecimiento, pero estos estaban concentrados en Plaza (a excepcion de una mipyme en la Habana Vieja). El proyecto de desarrollo local sin duda fue lo que mas llamo su atencion, ya que esta iniciativa se propone ayudar a mejora de esa localidad y para Carlos fue una pena no haber encontrado ningun otro local como este para su muestra.

Esto muestra que aunque estas bebidas son populares y consideradas parte de la cultura gastronomica cubana no todos los establecimientos se centran en su oferta sobre todo aquellos mas alejados de las zonas que podriamos considerar mas visitadas por los turistas. Con esto hablamos de la preferencia del cliente, los propietarios de los locales pueden adaptar su oferta a las preferencias de estos. Si observan que los cocteles no son populares entre los clientes pueden elegir no incluirlos en su menu. Municipios como San Miguel, Arroyo Naranjo y el Cotorro, los restaurantes de su muestra no ofrecian ninguno de los cuatro cocteles, y la mayor concentracion de locales que las ofertan se encuentra en La Habana Vieja.

Para profundizar en su investigacion sobre la disponibilidad de los cuatro cocteles, Carlos con la ayuda de la grafica anterior en donde analizo los restaurantes que ofertaban cocteles (La grafica color morado) y la segunda grafica (color azul) que presenta la cantidad de establecimientos en esos mismos municipios que ofertan en la carta los 4 cocteles. El proposito de este analisis fue determinar si existia una relacion entre la cantidad de restaurantes en un municipio y la probabilidad de encontrar un establecimiento que ofreciera todos los cocteles analizados. Al comparar las dos graficas, Carlos observo que existia una relacion entre ellas y es que concluyo que las probabilidades de encontrar un establecimiento que los ofertaran los 4 esta relacionada a la cantidad de restaurantes que tenga ese municipio. A medida que estos disminuyen, la posibilidad de que un lugar sirva todos los cocteles tambien lo hace.  

```python
from bibliotheque import Cantidad
from bibliotheque import Bebidas
colores = ["yellow", "lime", "green", "lawngreen"]
split = (0,0,0.1,0)
plt.pie(Cantidad, labels = Bebidas, autopct = "%0.1f%%", colors= colores, explode= split)
plt.axis("equal")
plt.show()
```

Una vez realizado el analisis de la muestra de restaurantes y bares que ofrecen los cuatro cocteles, Carlos concluyo que el mojito resulto ser el coctel mas ofertado, mientras que la piña colada tuvo el ultimo lugar en  frecuencia en los menus. Esto podria deberse a varios factores como que el mojito es un coctel emblematico de Cuba y muchos establecimientos pueden querer incluirlo en su carta como una bebida representativa del lugar. Ademas los ingredinetes del mojito como la menta, el ron y el azucar son relativamente faciles de conseguir. Por otro lado la piña colada aunque tambien es otra bebida popular, puede requerir de ingredientes mas costosos como la piña fresca y la crema de coco. Esto podria explicar parte de su poca frecuencia en los menus habaneros.
